```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import ascent
from scipy.signal import convolve2d
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import MaxPooling2D 
from tensorflow.keras.layers import AveragePooling2D

def showImages(orig, norm):
    plt.clf()
    fig, ax = plt.subplots(1,2,figsize=(10,3))
    original = ax[0].imshow(orig, cmap="gray")
    ax[0].set_title("Original")
    fig.colorbar(original, ax=ax[0])
    
    normImage = ax[1].imshow(norm, cmap="gray")
    fig.colorbar(normImage, ax=ax[1])
    ax[1].set_title("Neu")
    plt.show()
```

# Layer 1 - Feature Learning
Dieses und die beiden nächsten Notebooks enthalten eine genaue Beschreibung der Layer, die in einem CNN auftreten.<br>
In diesem wird die Konvolutionsschicht und die Poolingschicht erklärt.

## Conv2D
Zuerst zur Konvolution, dem zentralen Element eines CNNs. Die Überschrift dieses Kapitels ist zusammengesetzt aus "Conv" - Convolution und "2D" - die Konvolution wird auf Bilder, also 2D-Daten, angewendet.
### Input
Eine Konvolutionsschicht enthält zweidimensionale Daten mit Breite $b$ und Höhe $h$ als Input, der Feature Map genannt wird.
### Verarbeitung
Für die Schicht wird ein Kernel der Größe $n\times n$ gespeichert ($n\leq b, n\leq h$).<br>
Der Kernel wird über die Inputdaten geschoben und für jede Position wird die Summe der Produkte aus dem Wert des Kernels an diesem Pixel und dem Wert der Daten gebildet. Das Ergebnis ist der Wert des Outputpixels an dieser Stelle.
### Output
Es werden zweidimensionale Daten zurückgegeben. Die Breite dieser ist $b-n+1$ und die Höhe $h-n+1$. In alternativen Implementierungen ändert sich die Breite und Höhe nicht.
### Beispiel
Bei der Konvolution wird ein Kernel auf eine Feature Map angewandt.

Anschaulich wird bei der Konvolution das Inputbild (bzw. die Feature Map) hergenommen und der Kernel darüber geschoben. Für jeden Schritt wird der Kernel mit dem Teil des Bildes, den er bedeckt, multipliziert. Diese Werte werden addiert und ergeben den neuen Pixelwert.<br>
In der Abbildung oben ist eine beispielhafte Konvolution dargestellt. Der verwendete $3\times 3$-Kernel dient zur Detektion von horizontalen Kanten und wird auch <a href="https://de.wikipedia.org/w/index.php?title=Sobel-Operator&oldid=204090356">Sobel-Operator</a> genannt. Im Bild rechts haben die Bereiche, in denen es keine Änderung der Pixelwerte in vertikaler Richtung gibt, alle einen sehr kleinen Wert. Wenn der Kernel jedoch den Sprung in den Werten bedeckt (die Kante), dann erzeugt er einen sehr großen Ergebniswert.<br>
Der gezeigte Kernel ist nur beispielhaft zu sehen, weil dieser im Lernprozess verändert wird und so der Kernel je nach CNN, in dem sich der Layer befindet, verschieden aussehen wird. Es kann also nur schwer erkannt werden, dass ein Kernel zur Detektion von Kanten, etc. in der Feature Map dient. Das stimmt mit der allgemeinen Meinung überein, dass ein neuronales Netzwerk mehr als Blackbox als als Algorithmus zu betrachten ist.<br>
Außerdem fällt wahrscheinlich auf, dass die resultierende Feature Map kleiner als der Input der Schicht ist. Wenn dies unerwünscht ist, können verschiedene Techniken eingesetzt werden, damit die Dimension der Feature Map gleich bleibt. Die folgende Liste beschreibt Padding-Methoden, die genutzt werden, um Pixel außerhalb der Feature Map für die Konvolution nutzen zu können:
<ul>
    <li>Nähester Pixel: der äußerste Pixel wird für alle Pixel genommen, die außerhalb des Bildes liegen</li>
    <li>Spiegeln: der 3. Pixel außerhalb hat den Wert des 3. innerhalb der Grenze</li>
    <li>Wrapping: der Pixel wird von der anderen Seite der Feature Map genommen</li>
</ul>
Eine weitere Methode ist auch, den Kernel anzupassen, indem die Werte außerhalb der Grenzen ignoriert werden, aber dafür den restlichen eine größere Bedeutung zukommt.<br>
Ein weiterer Faktor bei der Konvolution ist die <b>Schrittweite</b>. In der Abbildung oben wird eine Schrittweite von $1$ verwendet, d. h. der Kernel wird in jedem Schritt einen Pixel weiter geschoben. Es sind aber auch größere Schrittweiten möglich. Bei diesen wird das Ergebnis kleinere Dimensionen haben als die ursprüngliche Feature Map.<br>
Ein letzter Punkt zur Konvolutionsschicht: die Schicht kann auch <b>mehrere Kernel</b> enthalten. Diese werden dann entweder als weitere Dimension der Feature Map weitergereicht, oder noch in der Schicht zusammengeführt.

## Code - Konvolution
Es folgt ein Beispiel, bei dem der oben genannte Sobel-Operator auf ein Bild angewandt wird. Im Bild rechts sind die waagerechten Kanten als Resultat deutlich hervorgehoben.<br>
Der erste Codeblock dient zur Erzeugung eines Kreises. Dann wird der Kernel definiert und die Konvolution durchgeführt. Im letzten Schritt wird das Bild des Kreises und das neue Bild angezeigt. Zur einfacheren Erkenntbarkeit wird der Absolutwert des Bildes genommen, eigentlich besteht die untere Hälfte des Rings aus negativen Werten.

```python
radius = 150
image = np.zeros((512, 512)).astype('uint8')
cx, cy = 256, 256
y, x = np.ogrid[-radius: radius, -radius: radius]
index = x**2 + y**2 <= radius**2
image[cy-radius:cy+radius, cx-radius:cx+radius][index] = 255

kernel = np.array([
    [1, 2, 1],
    [0, 0, 0],
    [-1,-2,-1]
])
plt.clf()
output = convolve2d(image, kernel)

showImages(image, np.abs(output))
```

## Pooling (Max und Avg)
### Input
Der Input in einen Pooling-Layer ist eine $b\times h$ Feature Map.
### Verarbeitung
Es wird wieder ein Teilbereich des Inputs betrachtet. Häufig hat der verwendete Kernel die Dimension $2\times 2$ ($n=2$). Kernel, bei denen sich die Breite und Höhe sich unterscheiden werden hier nicht betrachtet.<br>
Je nach verwendeter Funktion (Maximum oder Mittelwert) wird ein Wert für die Überlagerung von Kernel und Feature Map an jeder Stelle berechnet, und dieses Ergebnis als Outputpixel genommen.<br>
Der Unterschied zur Konvolution ist einerseits, dass der Kernel im Lernprozess (Backpropagation) nicht verändert wird, und andererseits, dass die Schrittweite größer ist. Der Kernel wird in jedem Schritt um mehr als einen Pixel verschoben, oft um die komplette Kernelbreite (/-höhe), sodass jeder Pixel nur einmal in die Berechnung des Outputs einfließt.<br>
#### Max Pooling
Beim Max Pooling wird für jede Überlagerung von Kernel und Feature Map nur der maximale Wert dieses Teilbereiches genommen. Das kann nicht als Matrix abgebildet werden, jedoch ist die Funktion trivial zu implementieren. Auf ein $2\times2$-Beispiel angewendet sieht das Max Pooling wie folgt aus

Die Backpropagation im Max Pooling wird jetzt erklärt. Beim forward pass wird eine Maske der Pixel gespeichert, die in den Output eingeflossen sind. Der Loss für die Pooling Schicht wird dann so aufgeteilt, dass nur die Pixel einen Loss bekommen, die auch von der Maske abgedeckt werden. Die folgende Animation dient zur besseren Verständlichkeit.

<h4>Avg Pooling</h4>
Beim Avg Pooling ist der Wert des Outputpixels gleich dem arithmetischen Mittelwert aller Werte des vom Kernel abgedeckten Teilbereiches. Diese Methode kann als Matrix abgebildet werden, mit allen Kernelwerten gleich $\frac{1}{n^2}$. Ein Beispiel:
<div style="width: 40% ; margin: auto;margin-top: 20px;margin-bottom: 20px;">
    <img src="assets/avgPoolEx.png" alt="Funktionsweise des Avg Poolings">
    <div style="text-align: center ; color: rgb(117 , 117 , 117)">Veranschaulichung des Avg Poolings</div>
</div>
Die Backpropagation funktioniert so, dass der Loss, der an einem Output-Pixel anliegt, gleichermaßen auf alle Input-Pixel (die in diesen Output-Pixel eingeflossen sind) aufgeteilt wird.
<h3>Output</h3>
Der Output ist eine Feature Map mit reduzieren Dimensionen. Für die Schrittweite $s$ in x- und y-Richtung gilt für die Dimension des Outputs:$$b_{out}=\frac{b_{in}}{s}, h_{out}=\frac{h_{in}}{s}$$

### Vergleich von Max und Avg Pooling

## Code - Pooling
Der folgende Code zeigt den Effekt eines Pooling Layers auf ein Bild. Es werden zuerst das Max Pooling gezeigt, dann das Avg Pooling.<br>
Auf den ersten Blick ist kein Unterschied zwischen dem Original und dem neuen Bild zu sehen. Doch haben die Bilder rechts die halbe Höhe und Breite. Das bedeutet eine Einsparung an Daten um $75\%$, die jedoch zu kaum Informationsverlust geführt hat.<br>
Außerdem ist es in diesem Fall irrelevant, ob das Max- oder das Avg-Pooling verwendet werden, das Ergebnis sieht sehr ähnlich aus.

```python
image = ascent()
arr = np.asfarray(image.reshape(1, image.shape[0], image.shape[1], 1))

# Max Pool
maxPool = MaxPooling2D(pool_size = 2, strides = 2)
modelMax = Sequential([maxPool])
imageMax = modelMax.predict(arr)
showImages(image, imageMax.reshape(imageMax.shape[1], imageMax.shape[2]))

# Avg Pool
avgPool = AveragePooling2D(pool_size = 2, strides = 2)
modelAvg = Sequential([avgPool])
imageAvg = modelAvg.predict(arr)
showImages(image, imageAvg.reshape(imageAvg.shape[1], imageAvg.shape[2]))
```

```python
fig = plt.figure(figsize=(3,1))
ax = fig.add_subplot(1,3,1)
prev = np.array([[1,2],[3,4]])
tb = ax.table(cellText=prev, loc=(0,0), cellLoc='center')

tc = tb.properties()['children']
for cell in tc: 
    cell.set_height(.5)
    cell.set_width(.5)

ax.set_xticks([])
ax.set_yticks([])

ax = fig.add_subplot(1,3,2)
ax.axis("off")
ax.set_xlim(0,1)
ax.set_ylim(-.5,.5)
ax.arrow(0,0,1,0, color="k", length_includes_head=True, head_width=.1, overhang=.75)

ax = fig.add_subplot(1,3,3)
after = np.squeeze(modelAvg.predict(prev.reshape(1, prev.shape[0], prev.shape[1], 1)),(0,3))
tb = ax.table(cellText=after, loc=(0,0), cellLoc='center')

tc = tb.properties()['children']
for cell in tc: 
    cell.set_height(.5)
    cell.set_width(.5)

ax.set_xticks([])
ax.set_yticks([])
ax.spines[:].set_visible(False)
fig.subplots_adjust(top=0.75)
fig.suptitle("Avg Pooling")
plt.savefig("avgPoolEx.png", transparent=False)
```
