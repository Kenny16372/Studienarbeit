# Bestimmte Hyperparameter erklärt
In diesem Notebook werden die in der Trainingsphase auftretenden Parameter **batch size**, **maximum number of iterations** und **validation split** erklärt.
## Batch size
Die **batch size** gibt an, wie viele Bilder in einem Durchlauf des CNN verwendet werden. Eine batch size von $1$ bedeutet, dass das Netzwerk mit jedem Bild einzeln trainiert wird, d. h. nach jedem Bild die Gewichte angepasst werden. Das Notebook zum Sequentiellen Modell beschreibt, wie genau das CNN lernt.
### Wie soll der Wert gewählt werden?
Die erste Punkt, der für den Parameter zu beachten ist, sind mögliche physikalische Grenzen.
1. Ein Wert $<1$ ist als Anzahl nicht zielführend
1. Genauso können auch keine Kommawerte angegeben werden
1. Nach oben ist die einzige Grenze der verfügbare Speicher. Wenn der Datensatz nicht komplett in den Speicher geladen werden kann, kann er auch nicht in einer Batch verarbeitet werden

Um zu entscheiden, ob eine batch size von $1$ oder $n$ (Anzahl an Datensätzen), oder aber ein Wert in der Mitte besser sind, werden die Vor- und Nachteile der beiden Extrema aufgelistet.
#### Vorteile Stochastischer Gradientenabstieg (batch size $=1$)
- Einzelne Schritte sind sehr schnell zu berechnen
- Können „schmalen“ Minima der Kostenfunktion entkommen
- Geringerer Speicherverbrauch

#### Nachteile Stochastischer Gradientenabstieg
- Viel Rauschen (Noise) in der Richtung der Gradienten
- Sehr viele Anpassungsvorgänge => großer Berechnungsaufwand
- Kann keine Vektorisierung nutzen

Die gerade genannten Stichpunkte werden jetzt erklärt. Bei dem Stochastischen Gradientenabstieg sind die Gradienten für die einzelnen Batches sehr schnell zu berechnen, da nur ein Bild pro Durchlauf verarbeitet wird. Das bedeutet aber auch, dass eine Epoche aus sehr vielen Durchläufen besteht. Der totale Berechnungsaufwand steigt also, weil die Anpassung der Gewichte nach jedem Durchlauf stattfindet. Um den Begriff der Epoche zu wiederholen: in einer Epoche werden alle Trainingsdatensätze einmal zum Training genutzt.<br>
Der nächste Punkt ist die Möglichkeit, dass „schmalen“ Minima der Kostenfunktion entkommen werden kann. Folgende Abbildung illustriert, warum das vorteilhaft ist.

In der vereinfachten Darstellung ist für einen Parameter die Kostenfunktion während des Trainings und während des Tests gezeigt. Die beiden Kurven sind gegeneinander verschoben, haben aber die gleiche Form. Die Form besteht aus einem schmalen und einem breiten Minimum. Man erkennt, dass beim schmalen Minimum der Trainings-Kostenfunktion der Wert der Test-Kostenfunktion viel höher ist als bei dem breiten Minimum. Daraus folgt, dass das breite Minimum stabiler ist und bei neuen Datensätzen vermutlich einen geringen Fehler hat als das schmale Minimum.<br>
Nun zu dem Argument, dass schmale Minima bei dem Stochastischen Gradientenabstieg vermieden werden können. Der Gradient eines einzelnen Bildes schwankt natürlicherweise mehr als der von Batches. Zur Veranschaulichung dient folgendes Codebeispiel, das die Mittelwerte von zufälligen Zahlen und Batches als Analogie hernimmt.

### Code - Varianz in Abhängigkeit der batch size
Der folgende Code erzeugt im ersten Schritt $100$ normalverteilte Zufallszahlen ($\mu = 0, \sigma^2 = 1$). Dann werden diese in $10$ Batches aufgeteilt und für jede Batch wird das arithmetische Mittel berechnet.<br>
Im nächsten Schritt wird ein Diagramm erstellt, das die Zufallszahlen (blau) und die Mittelwerte der Batches (gelb) darstellt.<br>
Zuletzt wird die Varianz zwischen einerseits den $100$ Werten und andererseits den $10$ Batches ausgegeben.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.clf()
randNums = np.random.randn((100))
batches = np.array_split(randNums, 10)
batchMean = [np.mean(batch) for batch in batches]
plt.plot(randNums)
plt.plot(np.linspace(0,100,10), batchMean)
plt.show()
print(f"Varianz bei batch size = 1: {np.var(randNums)}")
print(f"Varianz bei batch size = 10: {np.var(batchMean)}")
```

Die hohe Varianz bei kleinerer batch size kommt uns hier zugute. Beim Gradientenabstieg werden dadurch häufig Schritte in die falsche Richtung getan. Die falsche Richtung bedeutet hier eine Richtung, die nicht in Richtung des idealen &ndash; jedoch während der Laufzeit nicht bekannten &ndash; Gradienten, also nicht zum Minimum deutet. Diese Schritte sorgen bei einem schmalen Minimum dafür, dass es &ndash; um sich auf die Abbildung oben zu beziehen &ndash; wahrscheinlich ist, dass dieses Minimum beim Training verlassen wird. Bei breiten Minima hingegen ist es weniger wahrscheinlich, dass die zufälligen Schritte in die falsche Richtung aus dem Minimum führen. Somit *führt der Stochastische Gradientenabstieg wahrscheinlicher zu breiten als zu schmalen Minima*. Der einzige Nachteil ist, dass mehr Schritte getan werden müssen, um die in die falsche Richtung auszugleichen.<br>
#### Vorteile batch size = Anzahl Datensätze (batch mode)
- Schnell für kleine Datensätze
- Direkter Weg zum Minimum
- Nutzen von Vektorisierung

#### Nachteile batch mode
- Anpassung der Gewichte dauert lange
- Großer Speicherbedarf
- Gefahr von schmalen Minima
- Praxiserfahrung zeigt geringere Genauigkeit

Im batch mode werden die Gradienten über den gesamten Datensatz berechnet. Deswegen wird immer der beste Weg zum Minimum der Kostenfunktion gewählt. Das nachteilige daran ist, dass vom initialen Wert das nächste Minimum angestrebt wird und keine Möglichkeit besteht, andere Minima zu erreichen. Das ist in der folgenden Abbildung dargestellt.

Nun zum Hauptargument für den batch mode. Die <b>Vektorisierung</b> ist vielleicht manchen bekannt, trotzdem gibt es hier einen kurzen Überblick über die Funktionsweise.<br>

<h3>Fazit zur batch size</h3>
Beide Modi bringen spezielle Vorteile mit sich, aber diese kommen auch mit Nachteilen. Deswegen ist die allgemeine Meinung zur batch size, dass <b>Mini-Batches</b> verwendet werden sollen. Damit sind batch sizes in der Größenordnung $10$ bis $100$ gemeint. Außerdem, um die Vektorisierung komplett nutzen zu können, sollten die batch sizes ein Faktor von $2$ sein. Eine batch size von $32$ sollte für den Start genügen, aber größere wie $64$, $128$ oder $256$ sollten immer ausprobiert werden. Einzige Ausnahmen sind kleine Datensätze bis ca. $1000$ Werte, für die eine batch size gleich der Anzahl an Trainingsdatensätzen oft die beste Lösung ist.
<h2>Maximum number of iterations</h2>
Nun zu der maximum number of iterations. Inhaltlich bedeutet sie <i>die Anzahl an Epochen, die das Netzwerk trainiert</i>. Wählt man sie <b>zu klein</b>, führt dies zu <b>Underfitting</b>, <b>zu groß</b> und es führt zu <b>Overfitting</b>. Ein Beispiel am MNIST-Datensatz: Wenn Underfitting geschieht, dann wird z. B. die  $6$ oft als $8$ erkannt, da manche Schreibweisen der $8$ oben rechts eine Lücke haben und der Algorithmus diese deswegen als $6$ erkennt. Overfitting bedeutet in diesem Fall, dass zwar alle Trainingsbilder zu $99.5\%$ richtig erkannt werden, wenn man das CNN dann aber auf den Test- oder Validierungsdatensatz anwendet, dann erkennt das Netzwerk nicht mehr die kleinen Details, anhand derer es die Trainingsbilder unterschieden hat, was in einer reduzierten Genauigkeit von $70-80\%$ (oder sogar geringer) resultiert.

Folgender Graph wird oft hergenommen, um zu entscheiden, wann der Trainingsvorgang eines Netzwerks beendet werden soll.

Man sieht auf der x-Achse die Epochen aufgetragen und auf der y-Achse den Fehler ($=1-Genauigkeit$). Dann sind zwei Kurven eingezeichnet, eine für den Trainingsfehler und eine weitere für den Validierungsfehler. Der Trainingsfehler nimmt kontinuierlich ab, die Geschwindigkeit der Änderung lässt aber mit der Zeit nach. Der Validierungsfehler hat zu Beginn einen ähnlichen Verlauf wie der Trainingsfehler, macht aber ab einem bestimmten Zeitpunkt eine <b>Kehrtwende</b> und steigt wieder an. Genau dann ist es wahrscheinlich zum Overfitting gekommen, und man sollte das Training beenden. Der genaue Zeitpunkt ist nicht allzu wichtig, solange man sich in diesem ungefähren Bereich befindet, werden höchstwahrscheinlich die besten Ergebnisse in der Praxis erzielt. Der Trainingsfehler sollte mit geplottet werden, da so die Bestimmung des Endzeitpunktes vereinfacht wird.
<h2>Validation split</h2>
Der validation split beschreibt die Aufteilung des Gesamtdatensatzes in Trainings-, Test- und Validierungsdatensätze. Es folgt eine kurze Erklärung, welche Daten wann und wie genutzt werden.<br>
Vor dem Training werden die Testdaten vom Rest abgespalten. Dann, in jeder Epoche, wird ein Teil der restlichen Daten als Validierungsdaten verwendet und der Rest zum Training. Diese zweite Aufteilung kann sich auch zwischen Epochen ändern, sodass immer verschiedene Werte als Validierungsdaten genommen werden. Jetzt zum eigentlichen Training. Es werden die Trainingsdaten hergenommen und das Netzwerk lernt mit diesen. Dann, um zu überprüfen, ob Overfitting eingetreten ist, wird der Fehler des Validierungsdatensatzes berechnet. Dies wird so lange fortgeführt, bis die maximum number of iterations erreicht ist. Dann hat das Netzwerk fertig gelernt und die Gewichte werden eingefroren. Erst jetzt wird der Testdatensatz hergenommen und der Fehler in diesem berechnet. Da das Netzwerk die Daten noch nie gesehen hat, ist dieser Wert der beste Indikator dafür, ob Overfitting eingetreten ist. Der finale Fehler des Netzwerks wird dann im Testdatensatz ermittelt.
<h3>Welche Aufteilung sollte gewählt werden?</h3>
Die Aufteilung hängt davon ab, wie groß der Datensatz ist und welches Modell trainiert wird. Umso mehr Parameter es gibt, die erlernt werden müssen, desto größer sollte der Validierungsdatensatz gewählt werden. Defaultwerte z. B. von PyTorch sind $70\%$ Trainingsdaten, $20\%$ Validierungsdaten und $10\%$ Testdaten.
