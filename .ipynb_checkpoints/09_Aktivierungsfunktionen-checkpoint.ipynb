# Aktivierungsfunktionen
Dieses Notebook beschäftigt sich mit den Aktivierungsfunktionen Sigmoid, ReLU und Softmax. Die ersten beiden werden im Feature Learning bei Konvolutions-Layern genutzt. Softmax kommt im letzten Dense-Layer zum Einsatz, um den Outputvektor zu erhalten.

## Sigmoid
Die [Sigmoidfunktion](https://de.wikipedia.org/wiki/Sigmoidfunktion) ist definiert als: $$sig(t)=\frac{1}{1+e^{-t}}$$
Der Vorteil dieser Funktion ist, dass sie einfach differenzierbar ist, was zu einer kürzeren Trainingsdauer des Netzwerks führt. Die Ableitung ist: $$sig'(t)=sig(t)(1-sig(t))$$

## Code - Sigmoidfunktion
Der folgende Codeausschnitt zeigt die Sigmoidfunktion und deren Ableitung.

```python
X = np.linspace(-8, 8, 64)
Y = 1 / (1 + np.exp(-X))
YPrime = np.multiply(Y, (1-Y))
plt.clf()
fig, ax = plt.subplots()
ax.plot(X, Y, label="sig(t)")
ax.plot(X, YPrime, "--", label="sig'(t)")
ax.grid(True)
plt.legend(prop={'size': 12})
plt.show()
```

## ReLU
Die ReLU-Funktion (ausgeschrieben: rectified linear unit) ist definiert als: $$relu(x)=max(0,x)$$ 
Diese Funktion hat den Vorteil, dass die Berechnung der Funktion selbst sehr performant ist. Jedoch ist dafür die Ableitung nicht stetig und springt an der Stelle $x=0$ von $0$ auf $1$. Die Formel der Ableitung ist: $$relu'(x)=\begin{cases}
    1       & \quad \text{wenn } x>0\\
    0  & \quad \text{sonst}
  \end{cases}$$

## Code - ReLU
Der folgende Code erzeugt eine Grafik mit der ReLU-Funktion und deren Ableitung.

```python
X = np.linspace(-4, 4, 161)
Y = [max(0, x) for x in X]
YPrime = (X>0) * 1
plt.clf()
fig, ax = plt.subplots()
ax.plot(X, Y, label="relu(x)")
ax.plot(X, YPrime, "--", label="relu'(x)")
ax.grid(True)
plt.legend(prop={'size': 12})
plt.show()
```

## Softmax
Softmax ist eine Aktivierungsfunktion, die im letzten Dense-Layer genutzt wird. Die Formel für sie lautet:
$$\sigma(x)_i=\frac{e^{x_i}}{\sum_j{e^{x_j}}}$$
Es schließt sich aus der Formel, dass die $\sigma(x)_i$ strikt positiv sind, und dass die Summe aller $\sigma(x)_i$ gleich $1$ ist. Herleitung:
$$\sum_i{\frac{e^{x_i}}{\sum_j{e^{x_j}}}}=\frac{1}{\sum_j{e^{x_j}}}\sum_i{e^{x_i}}=\frac{\sum_i{e^{x_i}}}{\sum_j{e^{x_j}}}=1$$
Somit ist die Funktion dazu geeignet, um im letzten Layer eine Wahrscheinlichkeitsverteilung zu erhalten.<br>
Was noch vorteilhaft ist, große Werte im Vergleich zum Rest werden weiter vergrößert, während kleine Werte verkleinert werden.<br>
Softmax wird wegen dieser drei Eigenschaften im letzten Layer genutzt. In vorherigen Layern ist die Funktion weniger nützlich, da sie Ausreißer verstärkt darstellt und deswegen die Daten verzerrt.
## Code - Softmax
Der folgende Code zeigt den (zur Lesbarkeit gerundeten) Output der Softmax-Funktion für mehrere Inputvektoren. Die Summe aller errechneten Vektoren ergibt stets $1$, die Werte sind immer positiv, und größere Werte werden stärker skaliert.

```python
def softmax(X):
    return np.round(np.exp(X)/np.sum(np.exp(X)),2)

vec = [1,1,1,1]
print(f"{vec}->{softmax(vec)}")
vec = [9,9,9,9]
print(f"{vec}->{softmax(vec)}")
vec = [1,0,0,-1]
print(f"{vec}->{softmax(vec)}")
vec = [1,0,0,0]
print(f"{vec}->{softmax(vec)}")
vec = [7,0,0,0]
print(f"{vec}->{softmax(vec)}")
```

## Code - Vergleich der Aktivierungsfunktionen Relu und Sigmoid
Der folgende Abschnitt enthält das CNN zur Klassifizierung von MNIST. Einmal wird als Aktivierungsfunktion die Sigmoidfunktion genutzt, das andere Mal ReLU. Es wird die Performance der beiden CNNs anhand deren Genauigkeit und Trainingsdauer gemessen.

```python
import numpy as np
from keras.datasets.mnist import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = load_data() # Load data

def scaleBetweenMinus1And1(x):
    mean = np.mean(x)
    maxVal = np.max(x)
    minVal = np.min(x)
    return (x - mean)/(maxVal - minVal)

x_train = scaleBetweenMinus1And1(x_train) # Prepare data
x_test = scaleBetweenMinus1And1(x_test)

x_train = np.expand_dims(x_train, axis=3)
x_test = np.expand_dims(x_test, axis=3)

filter_count = 8
kernel_size = 3
pooling_size = 2

modelSigmoid = Sequential([ # Build Sigmoid model
  Conv2D(filter_count, kernel_size, input_shape=(28, 28, 1), activation='sigmoid'),
  MaxPooling2D(pool_size=pooling_size),
  Flatten(),
  Dense(10, activation='softmax'),
])

modelSigmoid.compile(
  'adam',
  loss='categorical_crossentropy',
  metrics=['accuracy'],
)

modelReLU = Sequential([ # Build ReLU model
  Conv2D(filter_count, kernel_size, input_shape=(28, 28, 1), activation='relu'),
  MaxPooling2D(pool_size=pooling_size),
  Flatten(),
  Dense(10, activation='softmax'),
])

modelReLU.compile(
  'adam',
  loss='categorical_crossentropy',
  metrics=['accuracy'],
)

histSigmoid = modelSigmoid.fit( # Train the Sigmoid model
  x_train,
  to_categorical(y_train),
  epochs=4,
  validation_data=(x_test, to_categorical(y_test)),
)

histReLU = modelReLU.fit( # Train the ReLU model
  x_train,
  to_categorical(y_train),
  epochs=4,
  validation_data=(x_test, to_categorical(y_test)),
)
```

## Ergebnis
Die ReLU-Funktion schneidet im Bezug auf die Accuracy besser ab. Außerdem benötigen beide eine ähnliche Dauer im Lernvorgang. Daraus folgt, dass ReLU als Aktivierungsfunktion im Konvolutionslayer zu bevorzugen ist.

```python
plt.clf()
plt.plot(histSigmoid.history['accuracy'], label="Sigmoid")
plt.plot(histReLU.history['accuracy'], label="ReLU")
plt.legend(prop={'size': 12})
plt.title("Accuracy im Vergleich von ReLU und Sigmoid")
plt.show()
```

```python
pred = modelReLU.predict(x_test) 
pred = np.argmax(pred, axis=1)[:10] 
label = y_test[:10] 
print(pred)
print(label)
```
